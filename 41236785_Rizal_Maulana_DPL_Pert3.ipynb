```python
# ============================================================
# FASHION-MNIST VAE – LATIHAN LENGKAP (EPOCHS = 50)
# 1. Membangun VAE sederhana untuk Fashion-MNIST
# 2. Visualisasi distribusi latent space
# 3. Latent space arithmetic (contoh pada Fashion-MNIST)
# 4. Perbandingan rekonstruksi antara VAE dan Autoencoder biasa
# 5. Sampling gambar baru dari latent prior
# 6. Perbandingan rekonstruksi loss VAE vs Autoencoder (angka)
# ============================================================

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

print("TensorFlow version:", tf.__version__)

# ------------------------------------------------------------
# 1. LOAD & PRE-PROCESS FASHION-MNIST
# ------------------------------------------------------------
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()

# Normalisasi ke [0,1]
x_train = x_train.astype("float32") / 255.0
x_test  = x_test.astype("float32") / 255.0

# Tambahkan channel dimension -> (batch, 28, 28, 1)
x_train = np.expand_dims(x_train, -1)
x_test  = np.expand_dims(x_test, -1)

input_shape = (28, 28, 1)
latent_dim = 2        # 2D agar mudah divisualisasikan

# BCE dengan reduction NONE supaya kita bisa kontrol sendiri
bce_loss_fn = keras.losses.BinaryCrossentropy(
    from_logits=False,
    reduction=tf.keras.losses.Reduction.NONE
)

# ------------------------------------------------------------
# 2. MEMBANGUN ENCODER
# ------------------------------------------------------------
encoder_inputs = keras.Input(shape=input_shape, name="encoder_input")
x = layers.Flatten()(encoder_inputs)
x = layers.Dense(512, activation="relu")(x)
x = layers.Dense(256, activation="relu")(x)

z_mean = layers.Dense(latent_dim, name="z_mean")(x)
z_log_var = layers.Dense(latent_dim, name="z_log_var")(x)

# Sampling layer: z = mean + exp(0.5 * log_var) * eps
class Sampling(layers.Layer):
    def call(self, inputs):
        z_mean, z_log_var = inputs
        epsilon = tf.random.normal(shape=tf.shape(z_mean))
        return z_mean + tf.exp(0.5 * z_log_var) * epsilon

z = Sampling()([z_mean, z_log_var])

encoder = keras.Model(encoder_inputs, [z_mean, z_log_var, z], name="encoder")
encoder.summary()

# ------------------------------------------------------------
# 3. MEMBANGUN DECODER
# ------------------------------------------------------------
latent_inputs = keras.Input(shape=(latent_dim,), name="z_sampling")
d = layers.Dense(256, activation="relu")(latent_inputs)
d = layers.Dense(512, activation="relu")(d)
d = layers.Dense(28 * 28, activation="sigmoid")(d)  # output flatten
decoder_outputs = layers.Reshape((28, 28, 1))(d)

decoder = keras.Model(latent_inputs, decoder_outputs, name="decoder")
decoder.summary()

# ------------------------------------------------------------
# 4. MEMBANGUN MODEL VAE (SUBCLASSING)
# ------------------------------------------------------------
class VAE(keras.Model):
    def __init__(self, encoder, decoder, **kwargs):
        super().__init__(**kwargs)
        self.encoder = encoder
        self.decoder = decoder

        self.total_loss_tracker = keras.metrics.Mean(name="loss")
        self.recon_loss_tracker = keras.metrics.Mean(name="recon_loss")
        self.kl_loss_tracker = keras.metrics.Mean(name="kl_loss")

    # dipakai kalau kita panggil vae(x)
    def call(self, inputs):
        z_mean, z_log_var, z = self.encoder(inputs)
        reconstruction = self.decoder(z)
        return reconstruction

    @property
    def metrics(self):
        # tracker untuk training
        return [
            self.total_loss_tracker,
            self.recon_loss_tracker,
            self.kl_loss_tracker,
        ]

    def compute_losses(self, data):
        """Hitung recon_loss, kl_loss, dan total_loss (dipakai train & test)."""
        z_mean, z_log_var, z = self.encoder(data)
        reconstruction = self.decoder(z)

        # Flatten ke (batch, 784)
        data_flat = tf.reshape(data, (tf.shape(data)[0], -1))
        recon_flat = tf.reshape(reconstruction, (tf.shape(reconstruction)[0], -1))

        # BCE per-sample: shape (batch,)
        per_sample_bce = bce_loss_fn(data_flat, recon_flat)
        recon_loss = tf.reduce_mean(per_sample_bce)

        # KL Divergence
        kl_loss = -0.5 * (1 + z_log_var - tf.square(z_mean) - tf.exp(z_log_var))
        kl_loss = tf.reduce_sum(kl_loss, axis=1)   # per-sample
        kl_loss = tf.reduce_mean(kl_loss)          # rata-rata batch

        total_loss = recon_loss + kl_loss
        return total_loss, recon_loss, kl_loss

    def train_step(self, data):
        if isinstance(data, tuple):   # (x, y) -> ambil x saja
            data = data[0]

        with tf.GradientTape() as tape:
            total_loss, recon_loss, kl_loss = self.compute_losses(data)

        grads = tape.gradient(total_loss, self.trainable_weights)
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))

        self.total_loss_tracker.update_state(total_loss)
        self.recon_loss_tracker.update_state(recon_loss)
        self.kl_loss_tracker.update_state(kl_loss)

        return {
            "loss": self.total_loss_tracker.result(),
            "recon_loss": self.recon_loss_tracker.result(),
            "kl_loss": self.kl_loss_tracker.result(),
        }

    def test_step(self, data):
        # dipakai untuk validation_data
        if isinstance(data, tuple):
            data = data[0]

        total_loss, recon_loss, kl_loss = self.compute_losses(data)

        # untuk validation, kita langsung return nilai saat ini
        return {
            "loss": total_loss,
            "recon_loss": recon_loss,
            "kl_loss": kl_loss,
        }

# buat dan compile VAE
vae = VAE(encoder, decoder)
vae.compile(optimizer=keras.optimizers.Adam())

# ------------------------------------------------------------
# 5. TRAINING VAE (EPOCHS = 50)
# ------------------------------------------------------------
history_vae = vae.fit(
    x_train,
    epochs=50,
    batch_size=128,
    validation_data=(x_test, None)
)

# ------------------------------------------------------------
# 6. VISUALISASI DISTRIBUSI LATENT SPACE (z_mean)
# ------------------------------------------------------------
# Hitung z_mean pada data test
z_mean_test, z_log_var_test, z_test = encoder.predict(x_test, batch_size=256, verbose=0)

plt.figure(figsize=(8, 6))
scatter = plt.scatter(
    z_mean_test[:, 0],
    z_mean_test[:, 1],
    c=y_test,
    cmap="tab10",
    s=4
)
plt.colorbar(scatter, ticks=range(10))
plt.xlabel("z[0]")
plt.ylabel("z[1]")
plt.title("Distribusi Latent Space Fashion-MNIST (z_mean)")
plt.grid(True)
plt.show()

fashion_labels = {
    0: "T-shirt/top",
    1: "Trouser",
    2: "Pullover",
    3: "Dress",
    4: "Coat",
    5: "Sandal",
    6: "Shirt",
    7: "Sneaker",
    8: "Bag",
    9: "Ankle boot",
}
print("Label kelas Fashion-MNIST:", fashion_labels)

# ------------------------------------------------------------
# 7. LATENT SPACE ARITHMETIC
#    Contoh: Sneaker (7) -> Ankle boot (9)
# ------------------------------------------------------------
CLASS_A = 7  # Sneaker
CLASS_B = 9  # Ankle boot

z_A = z_mean_test[y_test == CLASS_A]
z_B = z_mean_test[y_test == CLASS_B]

z_A_mean = np.mean(z_A, axis=0)
z_B_mean = np.mean(z_B, axis=0)

direction = z_B_mean - z_A_mean   # vektor perbedaan fitur

# Ambil beberapa contoh Sneaker
num_examples = 8
idx_A = np.where(y_test == CLASS_A)[0][:num_examples]
z_A_samples = z_mean_test[idx_A]

# alpha: tingkat "perpindahan" dari Sneaker ke Ankle boot
alphas = np.linspace(0, 1, 5)   # 0, 0.25, ..., 1

decoded_images = []

for i in range(num_examples):
    row_imgs = []
    for a in alphas:
        z_new = z_A_samples[i] + a * direction
        z_new = np.expand_dims(z_new, axis=0)
        x_dec = decoder.predict(z_new, verbose=0)
        row_imgs.append(x_dec[0, :, :, 0])
    decoded_images.append(row_imgs)

# Plot hasil latent space arithmetic
fig, axes = plt.subplots(num_examples, len(alphas),
                         figsize=(len(alphas)*2, num_examples*2))

for i in range(num_examples):
    for j, a in enumerate(alphas):
        ax = axes[i, j]
        ax.imshow(decoded_images[i][j], cmap="gray")
        ax.axis("off")
        if i == 0:
            ax.set_title(f"alpha={a:.2f}")

plt.suptitle("Latent Space Arithmetic: Sneaker -> Ankle boot", y=0.92)
plt.show()

# ------------------------------------------------------------
# 8. INTERPOLASI LATEN ANTAR DUA GAMBAR
#    Contoh: Dress (3) -> Coat (4)
# ------------------------------------------------------------
idx1 = np.where(y_test == 3)[0][0]  # Dress
idx2 = np.where(y_test == 4)[0][0]  # Coat

z1 = z_mean_test[idx1]
z2 = z_mean_test[idx2]

t_values = np.linspace(0, 1, 10)
imgs_interp = []

for t in t_values:
    zt = (1 - t) * z1 + t * z2
    x_dec = decoder.predict(np.expand_dims(zt, 0), verbose=0)
    imgs_interp.append(x_dec[0, :, :, 0])

plt.figure(figsize=(15, 2))
for i, img in enumerate(imgs_interp):
    ax = plt.subplot(1, len(imgs_interp), i + 1)
    ax.imshow(img, cmap="gray")
    ax.axis("off")
plt.suptitle("Interpolasi Latent Space: Dress -> Coat")
plt.show()

# ------------------------------------------------------------
# 9. MEMBANGUN AUTOENCODER BIASA UNTUK PERBANDINGAN
# ------------------------------------------------------------
ae_inputs = keras.Input(shape=input_shape, name="ae_input")
x = layers.Flatten()(ae_inputs)
x = layers.Dense(512, activation="relu")(x)
x = layers.Dense(256, activation="relu")(x)
latent_ae = layers.Dense(latent_dim, activation=None, name="latent_ae")(x)

d = layers.Dense(256, activation="relu")(latent_ae)
d = layers.Dense(512, activation="relu")(d)
d = layers.Dense(28 * 28, activation="sigmoid")(d)
ae_outputs = layers.Reshape((28, 28, 1))(d)

autoencoder = keras.Model(ae_inputs, ae_outputs, name="autoencoder")
autoencoder.summary()

autoencoder.compile(
    optimizer=keras.optimizers.Adam(),
    loss="binary_crossentropy"
)

# Training 50 epochs agar adil dengan VAE
history_ae = autoencoder.fit(
    x_train, x_train,
    epochs=50,
    batch_size=128,
    validation_data=(x_test, x_test)
)

# ------------------------------------------------------------
# 10. PERBANDINGAN REKONSTRUKSI:
#     ORIGINAL vs AUTOENCODER vs VAE
# ------------------------------------------------------------
n = 10  # jumlah contoh yang ditampilkan
test_samples = x_test[:n]

# Rekonstruksi Autoencoder
recon_ae = autoencoder.predict(test_samples, verbose=0)
# Rekonstruksi VAE
recon_vae = vae.predict(test_samples, verbose=0)   # pakai call()

plt.figure(figsize=(15, 5))

for i in range(n):
    # Original
    ax = plt.subplot(3, n, i + 1)
    ax.imshow(test_samples[i].reshape(28, 28), cmap="gray")
    ax.axis("off")
    if i == 0:
        ax.set_ylabel("Original")

    # Autoencoder
    ax = plt.subplot(3, n, i + 1 + n)
    ax.imshow(recon_ae[i].reshape(28, 28), cmap="gray")
    ax.axis("off")
    if i == 0:
        ax.set_ylabel("Autoencoder")

    # VAE
    ax = plt.subplot(3, n, i + 1 + 2*n)
    ax.imshow(recon_vae[i].reshape(28, 28), cmap="gray")
    ax.axis("off")
    if i == 0:
        ax.set_ylabel("VAE")

plt.suptitle("Perbandingan Rekonstruksi: Original vs Autoencoder vs VAE", y=0.95)
plt.show()

# ------------------------------------------------------------
# 11. PLOT HISTORY LOSS TRAINING VAE & AE
# ------------------------------------------------------------
plt.figure(figsize=(8, 4))
plt.plot(history_vae.history["loss"], label="VAE train loss")
plt.plot(history_vae.history["val_loss"], label="VAE val loss")
plt.legend()
plt.title("History Loss VAE")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 4))
plt.plot(history_ae.history["loss"], label="AE train loss")
plt.plot(history_ae.history["val_loss"], label="AE val loss")
plt.legend()
plt.title("History Loss Autoencoder Biasa")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True)
plt.show()

# ------------------------------------------------------------
# 12. SAMPLING GAMBAR BARU DARI PRIOR N(0, I)
# ------------------------------------------------------------
n_grid = 15  # ukuran grid: 15x15
grid_x = np.linspace(-3, 3, n_grid)
grid_y = np.linspace(-3, 3, n_grid)

figure = np.zeros((28 * n_grid, 28 * n_grid))

for i, yi in enumerate(grid_y):
    for j, xi in enumerate(grid_x):
        # sample titik z dari grid (xi, yi)
        z_sample = np.array([[xi, yi]], dtype="float32")
        x_dec = decoder.predict(z_sample, verbose=0)
        digit = x_dec[0, :, :, 0]
        # letakkan di canvas besar
        figure[i * 28 : (i + 1) * 28,
               j * 28 : (j + 1) * 28] = digit

plt.figure(figsize=(8, 8))
plt.imshow(figure, cmap="gray")
plt.axis("off")
plt.title("Sampling dari Latent Space (z ~ N(0, I))")
plt.show()

# ------------------------------------------------------------
# 13. PERBANDINGAN REKONSTRUKSI LOSS NUMERIK:
#     VAE vs AUTOENCODER DI DATA TEST
# ------------------------------------------------------------
# --- Rekonstruksi loss VAE di test set ---
total_vae_test, recon_vae_test, kl_vae_test = vae.compute_losses(x_test)
print("VAE - total loss (test):      ", float(total_vae_test.numpy()))
print("VAE - reconstruction loss:    ", float(recon_vae_test.numpy()))
print("VAE - KL loss:                ", float(kl_vae_test.numpy()))

# --- Rekonstruksi loss Autoencoder di test set ---
x_test_flat = x_test.reshape((x_test.shape[0], -1))
x_pred_ae = autoencoder.predict(x_test, verbose=0)
x_pred_ae_flat = x_pred_ae.reshape((x_pred_ae.shape[0], -1))

per_sample_bce_ae = bce_loss_fn(x_test_flat, x_pred_ae_flat)
recon_ae_test = tf.reduce_mean(per_sample_bce_ae)

print("Autoencoder - reconstruction loss (test):", float(recon_ae_test.numpy()))
```
